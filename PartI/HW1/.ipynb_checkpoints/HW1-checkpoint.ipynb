# Problem 1

```python
import numpy as np #numpy is only used for generate random number between (0, 1)
import matplotlib.pyplot as plt #matplotlib is only used for FE plot
```

```python
class ProblemOne:
    def __init__(self, start_x1, start_x2):
        self.init_x1 = start_x1
        self.init_x2 = start_x2
        self.best_val_list = list()
        self.end_count = 0
        self.max_iter = 100
    
    def solve_with_hill_climbing(self):
        count = 0
        best_x1 = self.init_x1
        best_x2 = self.init_x2
        best_val = self.get_objective_value(best_x1, best_x2)
        print(f"Initilized objective value is {best_val}")
        while count < self.max_iter:          
            best_x1_neigh, best_x2_neigh, best_val_neigh = self.get_best_neighbor(best_x1, best_x2)
            #CaseI: neighbor is better than current best value
            if (best_val_neigh <= best_val):
                best_x1 = best_x1_neigh
                best_x2 = best_x2_neigh
                best_val = best_val_neigh
            #CaseII: all neighbor objective values are smaller than current best value
            else:
                break
            count+=1
            print(f"Iteration {count}, x1 = {round(best_x1, 2)}, x2 = {round(best_x2, 2)}, best objective value = {round(best_val, 2)}")
            self.best_val_list.append(best_val)
        self.end_count = count

    def solve_with_random_walk(self):
        count = 0
        cur_x1 = self.init_x1
        cur_x2 = self.init_x2
        cur_val = self.get_objective_value(cur_x1, cur_x2)        
        best_x1 = cur_x1
        best_x2 = cur_x2
        best_val = cur_val
        while count < self.max_iter:          
            best_x1_neigh, best_x2_neigh, best_val_neigh = self.get_best_neighbor(cur_x1, cur_x2)
            if (best_val_neigh <= best_val):
                best_x1 = best_x1_neigh
                best_x2 = best_x2_neigh
                best_val = best_val_neigh
            cur_x1 = best_x1_neigh
            cur_x2 = best_x2_neigh
            cur_val = best_val_neigh
            count+=1
            print(f"Iteration {count}, x1 = {round(best_x1, 2)}, x2 = {round(best_x2, 2)}, best objective value = {round(best_val, 2)}")
            self.best_val_list.append(best_val)
        self.end_count = count

    def solve_with_simulated_annealing(self, init_temp, mutation_range):
        count = 1
        #The count that objective value didn't update
        patience = 0
        max_patience = 3
        #When the temperature is lower than temp_stop, the algorithm stops
        temp_stop = 1
        cur_x1 = self.init_x1
        cur_x2 = self.init_x2
        cur_val = self.get_objective_value(cur_x1, cur_x2)
        best_x1 = self.init_x1
        best_x2 = self.init_x2
        best_val = cur_val
        cur_temp = self.get_temp(count, init_temp)
        while count < self.max_iter and patience < max_patience and cur_temp > temp_stop:
            #print("="*50)
            update_threshold = self.get_uniform_dist()
            x1_neigh, x2_neigh, mutation_val = self.get_mutation_objective_value_SA(best_x1, best_x2, mutation_range)
            diff = mutation_val - cur_val
            update_prob = self.get_boltzmann_prob(diff, cur_temp)
            #print(diff, cur_temp, update_prob)
            if update_prob > update_threshold:
                #record best value
                if mutation_val < best_val:
                    best_x1 = x1_neigh
                    best_x2 = x2_neigh
                    best_val = mutation_val
                #update
                cur_x1 = x1_neigh
                cur_x2 = x2_neigh
                cur_val = mutation_val
                patience = 0
            else:
                patience += 1    
            count += 1
            print(f"Iteration {count}, current temperature = {round(cur_temp, 2)}, x1 = {round(best_x1, 2)}, x2 = {round(best_x2, 2)}, best objective value = {round(best_val, 2)}")
            self.best_val_list.append(best_val)
        self.end_count = count

    def get_best_neighbor(self, cur_x1, cur_x2):
        """test objectvie value of nine points around current working vector and return the best neighbor
        """
        bounds = [-0.1, 0, 0.1]
        best_val_neigh = 99999
        best_x1 = cur_x1
        best_x2 = cur_x2
        for step_x1 in bounds:
            for step_x2 in bounds:
                #Don't compare to itself
                if (step_x1 == 0 and step_x2 == 0):
                    continue
                cur_val_neigh = self.get_objective_value(cur_x1 + step_x1, cur_x2 + step_x2)
                if cur_val_neigh <= best_val_neigh:
                    best_val_neigh = cur_val_neigh
                    best_x1 = cur_x1 + step_x1
                    best_x2 = cur_x2 + step_x2                    
        return best_x1, best_x2, best_val_neigh

    def get_mutation_objective_value_SA(self, x1, x2, mutation_range):
        u1 = self.get_uniform_dist()
        u2 = self.get_uniform_dist()
        r1 = (x1 - mutation_range) + u1 * (mutation_range * 2)
        r2 = (x2 - mutation_range) + u2 * (mutation_range * 2)
        mutation_obj_val = self.get_objective_value(r1, r2)
        return r1, r2, mutation_obj_val

    def get_objective_value(self, x1, x2):
        value = 6 * (x1 ** 2) - 6 * x1 * x2 + 2 * (x2 ** 2) - x1 - 2 * x2
        return value
    
    def get_uniform_dist(self):
        """Get probability from a uniform distribution (0, 1)
        """
        return round(np.random.rand(), 2)
    
    def get_boltzmann_prob(self, delta_f, temperature):
        prob = min(1, np.exp(- delta_f / temperature))
        return prob
    
    def get_temp(self, num_iter, initial_temp):
        eplison = 0.05
        temp = ((1 - eplison) ** num_iter) * initial_temp
        return temp
```

```python
init_x1 = 3.42
init_x2 = 7.2
init_temp = 93.8
```

```python
P1 = ProblemOne(init_x1, init_x2)
P1.solve_with_hill_climbing()
iter_hill_climbing = [i+1 for i in range(P1.end_count)]
object_value_hill_climbing = P1.best_val_list
```

```python
P1 = ProblemOne(init_x1, init_x2)
P1.solve_with_random_walk()
iter_random_walk = [i+1 for i in range(P1.end_count)]
object_value_random_walk = P1.best_val_list
```

```python
range_of_mutation = 3
P1 = ProblemOne(init_x1, init_x2)
P1.solve_with_simulated_annealing(init_temp, range_of_mutation)
iter_SA = [i for i in range(1, P1.end_count)]
object_value_SA = P1.best_val_list
```

```python
plt.plot(iter_SA, object_value_SA, label = "Simulated Annealing")
plt.plot(iter_random_walk, object_value_random_walk, label = "Random Walk")
plt.plot(iter_hill_climbing, object_value_hill_climbing, label = "Hill Climbing")
plt.xlabel("# of iterations")
plt.ylabel("objective value")
plt.legend()
plt.show()
```

```python
distance_matrix = [
    [0, 27 ,335 , 244 , 141 , 257 , 33 , 316 , 186 , 115 , 304 , 439 , 102 , 95 , 275],
    [27 , 0 , 330 , 237 , 144 , 268 , 31 , 307 ,  195 , 113 , 301 , 453 , 75 , 111 , 290],
    [335 , 330 , 0 , 95 , 199 , 193 , 304 , 54 , 189 , 221 , 35 , 291 , 330 , 271 , 233],
    [244 , 237 , 95 , 0 , 117 , 171 , 212 , 75 , 130 , 130 , 72 , 324 , 236 , 191 , 215],
    [141 , 144 , 199 , 117 , 0 , 137 , 114 , 192 , 61 , 36 , 167 , 323 , 175 , 74 , 171],
    [257 , 268 , 193 , 171 , 137 , 0 , 238 , 222 , 77 , 173 , 161 , 186 , 311 , 162 , 44],
    [33 , 31 , 304 , 212 , 114 , 238 , 0 , 284 , 164 , 84 , 274 , 423 , 91 , 83 , 260],
    [316 , 307 , 54 , 75 , 192 , 222 , 284 , 0 , 198 , 205 , 67 , 341 , 296 , 266 , 265],
    [186 , 195 , 189 , 130 , 61 , 77 , 164 , 198 , 0 , 96 , 154 , 263 , 234 , 97 , 111],
    [115 , 113 , 221 , 130 , 36 , 173 , 84 , 205 , 96 , 0 , 190 , 359 , 139 , 74 , 205],
    [304 , 301 , 35 , 72 , 167 , 161 , 274 , 67 , 154 , 190 , 0 , 275 , 306 , 237 , 202],
    [439 , 453 , 291 , 324 , 323 , 186 , 423 , 341 , 263 , 359 , 275 , 0 , 498 , 344 , 165],
    [102 , 75 , 330 , 236 , 175 , 311 , 91 , 296 , 234 , 139 , 306 , 498 , 0 , 170 , 340],
    [95 , 111 , 271 , 191 , 74 , 162 , 83 , 266 , 97 , 74 , 237 , 344 , 170 , 0 , 180],
    [275 , 290 , 233 , 215 , 171 , 44 , 260 , 265 , 111 , 205 , 202 , 165 , 340 , 180 , 0]
]
```

```python
class TravelingSalesManProblem:
    def __init__(self, distance_matrix):
        self.distance_matrix = distance_matrix
        self.max_iter = 100
        self.max_tabu_length = 10
        self.best_route_length_list = list()

    def solve_with_random_walk(self):
        #Initialize a solution
        current_solution = self.generate_random_sol()
        current_route_length = self.get_route_length(current_solution)
        best_solution = current_solution
        best_route_length = current_route_length
        #Start solving
        count = 0
        update_threshold = 0.5
        while count < self.max_iter:
            neighbors = self.get_neighbors(current_solution)
            best_neighbor, best_neighbor_route_length = self.get_best_neighbor(neighbors)
            #updating best solution
            if best_neighbor_route_length < current_route_length:
                best_solution = best_neighbor
                best_route_length = best_neighbor_route_length

            #Must Update
            current_solution = best_neighbor
            current_route_length = best_neighbor_route_length
            count += 1
            self.best_route_length_list.append(best_route_length)
            print(f"Iteration {count}")
            print(f"Current solution is {current_solution}, current length is {current_route_length}")
            print(f"Best solution is {best_solution}, best length is {best_route_length}")
            print("=" * 50)

    def solve_with_hill_climbing(self):
        #Initialize a solution
        best_solution = self.generate_random_sol()
        best_route_length = self.get_route_length(best_solution)
        count = 0
        #Loop until it's not updating
        while count < self.max_iter:
            neighbors = self.get_neighbors(best_solution)
            best_neighbor, best_neighbor_route_length = self.get_best_neighbor(neighbors)
            if best_neighbor_route_length < best_route_length:
                best_solution = best_neighbor
                best_route_length = best_neighbor_route_length
            count += 1
            self.best_route_length_list.append(best_route_length)
            print(f"Iteration {count}, ", best_solution, f", best length is {best_route_length}")

    def solve_with_tabu_search(self):
        #Initialize a solution
        current_solution = self.generate_random_sol()
        current_route_length = self.get_route_length(current_solution)
        best_solution = current_solution
        best_route_length = current_route_length
        count = 0 
        tabu_list = []
        #Start solving
        while count < self.max_iter:
            neighbors = self.get_neighbors(current_solution)
            best_neighbor = None
            best_neighbor_route_length = None
            for neighbor in neighbors:
                neighbor_route_length = self.get_route_length(neighbor)
                update_condition = (neighbor_route_length <= best_route_length or ((neighbor not in tabu_list) and ((best_neighbor is None) or (neighbor_route_length < best_neighbor_route_length))))
                if update_condition is True:
                    best_neighbor = neighbor
                    best_neighbor_route_length = neighbor_route_length

            current_solution = best_neighbor
            current_route_length = best_neighbor_route_length
            if current_solution is not None:
                if current_route_length < best_route_length:
                    best_route_length = current_route_length
                    best_solution = current_solution
                #even if best neighbor is larger than best solution, we still append it to tabu list    
                tabu_list.append(current_solution)     
                if len(tabu_list) > self.max_tabu_length:
                    tabu_list = tabu_list[:self.max_tabu_length]

            count += 1
            #print(best_neighbor, tabu_list)
            self.best_route_length_list.append(best_route_length)
            #print(best_neighbor in tabu_list)
            print(f"Iteration {count}, ", best_solution, f", best length is {best_route_length}")

    def solve_with_simulated_annealing(self, init_temp):
        count = 0 #Number of iteration
        patience = 0 #The count that objective value didn't update
        max_patience = 3
        temp_stop = 1 #When the temperature is lower than temp_stop, the algorithm stops
        #Initialize a solution
        cur_solution = self.generate_random_sol()
        cur_route_length = self.get_route_length(cur_solution)
        best_solution = cur_solution
        best_route_length = cur_route_length
        cur_temp = self.get_temp(count, init_temp)
        #Stopping condition
        #When the objective value did't update for `max_patience` times OR
        #The temperature is lower than `temp_stop` OR
        #number of iterations is larger than 100
        while count < self.max_iter and patience < max_patience and cur_temp > temp_stop:
            update_threshold = self.get_uniform_dist()
            count+=1
            cur_temp = self.get_temp(count, init_temp)
            #Generating best neighbor solutions
            neighbors = self.get_neighbors(best_solution)
            best_neighbor, best_neighbor_route_length = self.get_best_neighbor(neighbors)
            #Calculate the difference between current (global) best solution
            #with best neighbor solution (local)
            diff = best_neighbor_route_length - cur_route_length
            update_prob = self.get_boltzmann_prob(diff, cur_temp)
            if update_prob > update_threshold:
                if best_neighbor_route_length < best_route_length:
                    best_solution = best_neighbor
                    best_route_length = best_neighbor_route_length
                #update
                cur_solution = best_neighbor
                cur_route_length = best_neighbor_route_length
                patience = 0
            else:
                patience += 1          
            self.best_route_length_list.append(best_route_length)
            print(f"Iteration {count}", best_solution, f", best length is {best_route_length}")

    def generate_random_sol(self):
        left_cities = [i for i in range(1, len(self.distance_matrix))] #Exclude Incheon
        sol = [] 
        for i in range(len(self.distance_matrix) - 1):
            rand_city_idx = np.random.randint(len(left_cities))
            rand_city = left_cities[rand_city_idx]
            sol.append(rand_city)
            left_cities.remove(rand_city)
        return sol
    
    def get_route_length(self, sol):
        ttl_length = 0
        #Add the distance from Incheon to sol[0]
        ttl_length += self.distance_matrix[0][sol[0]]
        for i in range(len(sol) - 1):
            ttl_length += self.distance_matrix[sol[i]][sol[i+1]]
        #Add the distance from sol[-1] to Incheon
        ttl_length += self.distance_matrix[sol[-1]][0]
        return ttl_length

    def get_neighbors(self, sol):
        """Generate all possible neighbor solutions by swapping elements in the original solution
        """
        neighbors = []
        for i in range(len(sol)):
            for j in range(i+1, len(sol)):
                neighbor = sol.copy()
                neighbor[i] = sol[j]
                neighbor[j] = sol[i]
                neighbors.append(neighbor)
        return neighbors

    def get_best_neighbor(self, neighbors):
        best_neighbor_route_length = self.get_route_length(neighbors[0])
        best_neighbor = neighbors[0]
        for neighbor in neighbors:
            current_route_length = self.get_route_length(neighbor)
            if current_route_length < best_neighbor_route_length:
                best_neighbor_route_length = current_route_length
                best_neighbor = neighbor
        return best_neighbor, best_neighbor_route_length

    def get_uniform_dist(self):
        """Get probability from a uniform distribution (0, 1)
        """
        return round(np.random.rand(), 2)
    
    def get_boltzmann_prob(self, delta_f, temperature):
        prob = min(1, np.exp(- delta_f / temperature))
        return prob
    
    def get_temp(self, num_iter, initial_temp):
        eplison = 0.05
        temp = ((1 - eplison) ** num_iter) * initial_temp
        return temp
```

```python
TSP = TravelingSalesManProblem(distance_matrix)
TSP.solve_with_hill_climbing()
route_length_hill_climbing = TSP.best_route_length_list
```

```python
TSP = TravelingSalesManProblem(distance_matrix)
TSP.solve_with_random_walk()
route_length_random_walk = TSP.best_route_length_list
```

```python
TSP = TravelingSalesManProblem(distance_matrix)
TSP.solve_with_tabu_search()
route_length_tabu = TSP.best_route_length_list
```

```python
x = list(range(1, TSP.max_iter + 1))
plt.plot(x, route_length_hill_climbing, label = "Hill Climbing")
plt.plot(x, route_length_random_walk, label = "Random Walk")
plt.plot(x, route_length_tabu, label = "Tabu Search")
plt.xlabel("# of iterations")
plt.ylabel("Best route length")
plt.legend()
plt.show()
```
