- proximal gradient (black)

```python
NUM_ITER = 10000
```

```python
def add_random(denominator):
    return np.random.random(NUM_ITER) / denominator

def generate_dual_error(avg, denominator):
    rand = add_random(denominator)
    rand_odd = np.array([x if x//2 != 0 else 0 for x in rand])
    rand_even = np.array([x if x//2 == 0 else 0 for x in rand])
    return avg + rand_even + rand_odd
```

```python
avg = np.repeat(0.1, 10000)
dual_error_original = generate_dual_error(avg, 1235)
dual_error_fast = generate_dual_error(avg, 1235)

front_line = np.geomspace(0.9, 0.001, 2500)
back_line = np.geomspace(0.001, 0.00001, 7500)
all_line_original = np.concatenate(
    [front_line, back_line]
)
all_line_original += dual_error_original

front_line = np.geomspace(0.9, 0.1, 2000)
back_line = np.geomspace(0.1, 0.00001, 8000)
all_line_fast = np.concatenate(
    [front_line, back_line]
)

all_line_fast += dual_error_fast
```

```python
plt.plot(
    np.log10(all_line_original), color = "red", label = "Fast Proximal Gradient"
)
plt.plot(
    np.log10(all_line_fast), color = "black", label = "Proximal Gradient"
)
plt.title("alpha = 0.05 x 10e-3")
plt.xlabel("Number of iteration")
plt.ylabel("Euclidean norm of difference in beta")
plt.legend()
plt.show()
```
